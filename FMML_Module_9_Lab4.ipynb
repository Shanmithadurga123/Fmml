## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

## Questions:
1) What is the significance of using data augmentations like resize, crop etc on training data?

2) What performance do you get if you don't use pretrained resnet model (Hint: Change pretrained=False and train the model)

3) If the resnet model was pre-trained on dataset significantly different than the ants vs bees data, would you still get good performance by using this pretrained model?


ANSWERS:

1. Data augmentation techniques like resize, crop, rotation, flipping, and others play a crucial role in machine learning, particularly in tasks like image classification, object detection, and natural language processing. Here's why they're significant:

1. **Increased Data Variability**: Augmentations introduce variations to the training data, making the model more robust and less sensitive to small changes in the input during inference. For example, resizing or cropping an image changes its aspect ratio, which helps the model generalize better to unseen data with different aspect ratios.

2. **Regularization**: Data augmentation acts as a form of regularization by adding noise to the training data, which can help prevent overfitting. It encourages the model to learn more generalizable features rather than memorizing the training examples.

3. **Improved Generalization**: By exposing the model to diverse variations of the input data, augmentation helps it learn more invariant and discriminative features. This leads to better generalization performance on unseen data.

4. **Balancing Classes**: In classification tasks with imbalanced classes, augmentations can be applied more frequently to the minority class samples, effectively balancing the distribution of classes in the training data.

5. **Reduced Dependency on Annotated Data**: Augmentations allow the generation of synthetic data, reducing the reliance on manually annotated data, which can be expensive and time-consuming to acquire.

6. **Domain Adaptation**: In transfer learning scenarios where the source and target domains have differences, data augmentation can be used to bridge the domain gap by simulating the variations present in the target domain.

7. **Better Handling of Input Variability**: Augmentations can help the model become more robust to real-world variations such as changes in lighting conditions, viewpoint, occlusions, and deformations.

Overall, data augmentation is a powerful technique for improving the performance and robustness of machine learning models, enabling them to perform better in real-world scenarios with diverse input data.

2. If you choose not to use a pretrained ResNet model in a machine learning task, the performance of your model may be affected in several ways:

1. **Training from Scratch**: Without using a pretrained model like ResNet, you would typically start with randomly initialized weights for your neural network. This means that your model needs to learn all the features from scratch during training, which can require a larger dataset and more computational resources.

2. **Limited Capacity for Feature Learning**: Pretrained models like ResNet have been trained on large datasets (e.g., ImageNet) and have learned a diverse set of features that are useful for a wide range of tasks. Without leveraging these pretrained features, your model may have a limited capacity to learn discriminative features, especially if your dataset is small or lacks diversity.

3. **Risk of Overfitting**: Training a deep neural network from scratch on a small dataset can lead to overfitting, where the model learns to memorize the training data rather than generalize to unseen data. This is especially true for complex models like ResNet, which have a large number of parameters.

4. **Increased Training Time**: Training a deep neural network from scratch can be computationally expensive and time-consuming, especially if you don't have access to powerful hardware or distributed computing resources.

5. **Potential for Suboptimal Performance**: Without leveraging the knowledge encoded in pretrained models, your model may not achieve the same level of performance or generalization as models that do use pretrained weights. This could result in suboptimal performance on your task of interest.

In summary, while it's possible to train a neural network from scratch without using a pretrained ResNet model, doing so may require more data, more computational resources, and may result in suboptimal performance compared to leveraging pretrained features.

3. Using a ResNet model pretrained on a significantly different dataset than the ants vs bees data may still provide benefits, but the extent of those benefits could vary depending on several factors:

1. **Transferability of Features**: Pretrained models, even if trained on a different dataset, often learn general features that are useful across a wide range of tasks. Lower layers of convolutional neural networks (CNNs) tend to capture low-level features like edges, textures, and basic shapes, which can be transferable across datasets. Higher layers capture more abstract and task-specific features, which may be less transferable.

2. **Domain Similarity**: The extent to which the source dataset (the one on which the ResNet was pretrained) is similar to the target dataset (ants vs bees) matters. If the two datasets share some underlying characteristics (e.g., both contain natural images), features learned by the pretrained model might still be relevant. However, if the datasets are very dissimilar (e.g., one is medical images while the other is satellite images), transferability might be limited.

3. **Fine-Tuning**: Even if the pretrained ResNet model was trained on a significantly different dataset, you can still fine-tune the model on the ants vs bees dataset. Fine-tuning involves updating the weights of the pretrained model using the target dataset. This allows the model to adapt its learned features to better suit the target task and dataset.

4. **Data Augmentation**: Data augmentation techniques can also help mitigate the differences between the source and target datasets. By applying transformations like rotation, flipping, and cropping to the images in the ants vs bees dataset, you can make the data more similar to the data the ResNet model was pretrained on.

In conclusion, while using a ResNet model pretrained on a significantly different dataset may not provide as much benefit as using a model pretrained on a more similar dataset, it can still be advantageous, especially when combined with fine-tuning and data augmentation techniques. However, the ultimate performance will depend on the specifics of the datasets and the task at hand.
